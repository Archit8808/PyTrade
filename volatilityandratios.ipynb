```python
import datetime as dt 
import pandas as pd 
import numpy as np 
from pandas_datareader import data as pdr 
import plotly.offline as pyo 
from plotly.subplots import make_subplots

pyo.init_notebook_mode(connected= True) 
pd.options.plotting.backend = 'plotly' 
```

```python
end = dt.datetime.now() 
start = dt.datetime(2021,6,14)

df = pdr.get_data_yahoo('PYPL',start,end) 
Close = df.Close 
Close.tail()
```

```python
log_returns = np.log(df.Close/df.Close.shift(1)).dropna()
log_returns 
```

```python
stdev = log_returns.std() 
stdev 

```

```python
annual_volitality = stdev *np.sqrt(252)
annual_volitality * 100
```

```python
#histogram of log returns against annual volitality 

fig = make_subplots(rows = 2 , cols = 2) 

log_returns.plot(kind = 'hist').update_layout(autosize = False , width = 500 , height = 300, title = "Log return frequency" , xaxis = dict(title = 'PYPL annual vol : '+str(np.round(annual_volitality * 100,1)))) 
```

```python
#trading volitality over time 

trade_days = 30
volitality = log_returns.rolling(window = trade_days).std()*np.sqrt(trade_days)
volitality
#the volitality formula is rho(T) = rho*sqrt(time period) 
# now r,rho would be the standard deviation of the log returns over time .
```

```python
volitality.plot()
```

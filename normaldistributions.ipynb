```python
import datetime as dt 
import pandas as pd 
import numpy as np 
import pylab 
import seaborn as sns 
import scipy.stats as stats 
from pandas_datareader import data as pdr 
import plotly.offline as pyo 

pyo.init_notebook_mode(connected= True) 
pd.options.plotting.backend = 'plotly' 
```

```python
end = dt.datetime(2022,6,13)
start = dt.datetime(2021,11,8) 

df = pdr.get_data_yahoo('PYPL' , start , end) 
df.head() 
```

```python
simple_returns = df.Close.pct_change().dropna()
simple_returns
```

```python
#multi period returns 

df.Close.plot().update_layout(autosize = False , width = 500 , height = 300) 
```

```python
print('First' , df.Close[0] , 'Last' , df.Close[-1])
```

```python
simple_returns.mean() #avg returns
```

```python
df.Close[0] * np.prod([(1 + Rt) for Rt in simple_returns])
```

```python
#log returns 

log_returns = np.log(df.Close/df.Close.shift(1)).dropna() 
log_returns
```

```python
log_returns.mean()
```

```python
df.Close[0] * np.exp(log_returns.mean()* len(log_returns)) 
```

```python
log_returns.plot(kind = 'hist').update_layout(autosize = False , width = 500 , height = 300) 
```

```python

#kolmogorov smirnov test to check if its a normal distribution by calculating the distance between the empirical(real) and theoretical distributions(ideal)
ks_stat , p_value = stats.kstest(log_returns , 'norm')
print(ks_stat,p_value)

if p_value > 0.05 : 
    print("Gaussian")
else : 
    print("not gaussian")
```
